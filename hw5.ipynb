 Create these specific entangled and product states, then verify using the separability test.

```python
def is_entangled(statevector):
    """Test if a 2-qubit state is entangled using αδ = βγ criterion"""
    sv = np.array(statevector).flatten()
    alpha, beta, gamma, delta = sv[0], sv[1], sv[2], sv[3]
    return not np.isclose(alpha * delta, beta * gamma, atol=1e-10)

# a) Create a PRODUCT state: |+⟩ ⊗ |-⟩ = (|00⟩ - |01⟩ + |10⟩ - |11⟩)/2
qc_product = QuantumCircuit(2)
# TODO: Add gates to create |+⟩ ⊗ |-⟩
qc_product.h(0)    # q0 in |+>
qc_product.x(1)    # q1 in |1>
qc_product.h(1)    # q1 in |->

sv = Statevector(qc_product)
print(f"a) State: {np.round(sv.data, 3)}")
print(f"   Expected: [0.5, -0.5, 0.5, -0.5]")
print(f"   Entangled? {is_entangled(sv)} (should be False)")
```

 ## Part 2: Bell State Properties (20 points)

 ### Exercise 2.1 (7 points)

 Measure Bell states and observe correlation patterns.

```python
def measure_correlations(qc, shots=1000):
    """Measure a circuit and return correlation E = P(same) - P(different)"""
    sv = Statevector(qc)
    counts = sv.sample_counts(shots=shots)
    same = counts.get('00', 0) + counts.get('11', 0)
    diff = counts.get('01', 0) + counts.get('10', 0)
    return (same - diff) / shots, counts

# Measure |Φ+⟩
qc1 = QuantumCircuit(2)
qc1.h(0)
qc1.cx(0, 1)
E1, counts1 = measure_correlations(qc1)
print(f"|Φ+⟩ correlation: {E1:.3f}, counts: {counts1}")

# Measure |Ψ+⟩
qc2 = QuantumCircuit(2)
qc2.x(1)
qc2.h(0)
qc2.cx(0, 1)
E2, counts2 = measure_correlations(qc2)
print(f"|Ψ+⟩ correlation: {E2:.3f}, counts: {counts2}")
```

 **Questions:**



 1. What outcomes do you see for |Φ+⟩? What outcomes do you NEVER see?
it always gives `00` or `11` , The qubits always match.


 2. What outcomes do you see for |Ψ+⟩? What outcomes do you NEVER see?
it always gives `01` or `10`, The qubits always differ.



 3. Both have |E| = 1 (perfect correlation). What's the difference?
|Φ+⟩ has same-outcome correlation (both 0 or both 1), while |Ψ+⟩ has opposite-outcome correlation (one 0, one 1).

```python
def measure_at_angles(theta_a, theta_b, shots=1000):
    """Create |Φ+⟩ and measure with rotated bases"""
    qc = QuantumCircuit(2)
    qc.h(0)
    qc.cx(0, 1)
    qc.ry(-theta_a, 0)  # Alice's measurement angle
    qc.ry(-theta_b, 1)  # Bob's measurement angle
    
    sv = Statevector(qc)
    counts = sv.sample_counts(shots=shots)
    same = counts.get('00', 0) + counts.get('11', 0)
    diff = counts.get('01', 0) + counts.get('10', 0)
    return (same - diff) / shots

# TODO: Fill in the measured correlations
print("Angle_A  Angle_B  Measured_E  Theory cos(θa-θb)")
print("-" * 50)

# Both measure in Z-basis
E = measure_at_angles(0, 0)
print(f"  0°       0°      {E:+.3f}       {np.cos(0):+.3f}")

# Alice Z, Bob at 45°
E = measure_at_angles(0, np.pi/4)
print(f"  0°      45°      {E:+.3f}       {np.cos(-np.pi/4):+.3f}")

# Alice Z, Bob X (90°)
E = measure_at_angles(0, np.pi/2)
print(f"  0°      90°      {E:+.3f}       {np.cos(-np.pi/2):+.3f}")

# Both at 45°
E = measure_at_angles(np.pi/4, np.pi/4)
print(f" 45°      45°      {E:+.3f}       {np.cos(0):+.3f}")

# Alice Z, Bob opposite (180°)
E = measure_at_angles(0, np.pi)
print(f"  0°     180°      {E:+.3f}       {np.cos(-np.pi):+.3f}")
```

 ## Part 3: Bell Measurements (20 points)

 ### Exercise 3.1 (7 points)

 Implement Bell measurement to distinguish the four Bell states.

```python
def bell_measure(qc):
    """Apply Bell measurement circuit and return result"""
    qc_m = qc.copy()
    qc_m.cx(0, 1)  # CNOT
    qc_m.h(0)      # Hadamard
    sv = Statevector(qc_m)
    counts = sv.sample_counts(shots=100)
    return list(counts.keys())[0]

# Test on all four Bell states
print("Bell State → Measurement Result")
print("-" * 35)

# |Φ+⟩
qc = QuantumCircuit(2)
qc.h(0); qc.cx(0, 1)
print(f"|Φ+⟩ → {bell_measure(qc)} (expected: 00)")

# |Φ-⟩
qc = QuantumCircuit(2)
qc.x(0); qc.h(0); qc.cx(0, 1)
print(f"|Φ-⟩ → {bell_measure(qc)} (expected: 01)")

# |Ψ+⟩
qc = QuantumCircuit(2)
qc.x(1); qc.h(0); qc.cx(0, 1)
print(f"|Ψ+⟩ → {bell_measure(qc)} (expected: 10)")

# |Ψ-⟩
qc = QuantumCircuit(2)
qc.x(0); qc.x(1); qc.h(0); qc.cx(0, 1)
print(f"|Ψ-⟩ → {bell_measure(qc)} (expected: 11)")
```

 **Question:** Why does the Bell measurement circuit (CNOT then H) distinguish all four Bell states?
It is the reverse of the creation circuit (which is H then CNOT). Applying the reverse maps each Bell state back to a unique computational basis state: |Φ+⟩→|00⟩, |Φ-⟩→|01⟩, |Ψ+⟩→|10⟩, |Ψ-⟩→|11⟩. Each result is distinct, so we can tell all four apart.

 Implement superdense coding: send 2 classical bits using 1 qubit + shared entanglement.

```python
def superdense_coding(bits):
    """
    Encode 2 classical bits and decode using Bell measurement.
    
    Protocol:
    1. Create shared |Φ+⟩ (Alice has qubit 0, Bob has qubit 1)
    2. Alice encodes her 2 bits by applying gates to qubit 0
    3. Bob performs Bell measurement to decode
    """
    qc = QuantumCircuit(2)
    
    # Step 1: Create shared Bell state
    qc.h(0)
    qc.cx(0, 1)
    qc.barrier()
    
    # Step 2: Alice encodes (applies gates to qubit 0 only)
    # TODO: Fill in the encoding
    if bits == '00':
        pass  # I (identity)
    elif bits == '01':
        qc.z(0)  # TODO: What gate?
    elif bits == '10':
        qc.x(0)  # TODO: What gate?
    elif bits == '11':
        qc.x(0)    
        qc.z(0)  
    
    qc.barrier()
    
    # Step 3: Bob performs Bell measurement
    qc.cx(0, 1)
    qc.h(0)
    
    return qc

# Test all 4 messages
print("Sent → Received")
print("-" * 20)
for bits in ['00', '01', '10', '11']:
    qc = superdense_coding(bits)
    sv = Statevector(qc)
    result = list(sv.sample_counts(100).keys())[0]
    match = "✓" if result == bits else "✗"
    print(f" {bits}  →   {result}    {match}")
```

 ### Exercise 3.3 (6 points)

 Explore how errors affect Bell states and measurements.

```python
def apply_error_and_measure(error_gate):
    """Create |Φ+⟩, apply error, then Bell measure"""
    qc = QuantumCircuit(2)
    qc.h(0)
    qc.cx(0, 1)  # |Φ+⟩
    
    # Apply error to qubit 0
    if error_gate == 'X':
        qc.x(0)
    elif error_gate == 'Z':
        qc.z(0)
    elif error_gate == 'Y':
        qc.y(0)
    
    # Bell measurement
    qc.cx(0, 1)
    qc.h(0)
    
    sv = Statevector(qc)
    return list(sv.sample_counts(100).keys())[0]

print("Error on |Φ+⟩ → Bell measurement result")
print("-" * 40)
print(f"No error  → {apply_error_and_measure(None)} (|Φ+⟩)")
print(f"X error   → {apply_error_and_measure('X')} (which Bell state?)")
print(f"Z error   → {apply_error_and_measure('Z')} (which Bell state?)")
print(f"Y error   → {apply_error_and_measure('Y')} (which Bell state?)")
```

 **Questions:**



 1. An X error on |Φ+⟩ transforms it to which Bell state?
An X error on |Φ+⟩ flips q0, giving (|10⟩ + |01⟩)/√2 = |Ψ+⟩. The Bell measurement returns 10.


 2. A Z error on |Φ+⟩ transforms it to which Bell state?
A Z error on |Φ+⟩ flips the sign of the |1⟩ component, giving (|00⟩ - |11⟩)/√2 = |Φ-⟩. The Bell measurement returns 01.


 3. How could this be used for error detection?
Each error maps |Φ+⟩ to a different Bell state, which gives a different unique measurement result. So by looking at the Bell measurement outcome, we can detect and identify which error occurred.

 ## Part 4: Challenge Problem (10 points)

 ### Challenge: Entanglement Swapping

 Create entanglement between qubits that have NEVER directly interacted!

```python
# Setup: 4 qubits in two separate Bell pairs
# Pair A: qubits 0 and 1 (entangled)
# Pair B: qubits 2 and 3 (entangled)
# 
# Goal: Entangle qubits 0 and 3 (which never interacted!)
# Method: Perform Bell measurement on qubits 1 and 2

qc = QuantumCircuit(4)

# Create first Bell pair (qubits 0, 1)
qc.h(0)
qc.cx(0, 1)

# Create second Bell pair (qubits 2, 3)
qc.h(2)
qc.cx(2, 3)

qc.barrier()

# TODO: Perform Bell measurement on qubits 1 and 2
# (This "swaps" the entanglement to qubits 0 and 3)
# Bell measurement on qubits 1 and 2
qc.cx(1, 2)   # CNOT: control=q1, target=q2
qc.h(1)       # Hadamard on q1

print("Circuit:")
print(qc.draw('text'))

# After Bell measurement on 1,2: qubits 0,3 become entangled!
sv = Statevector(qc)
print("\nStatevector after Bell measurement on qubits 1,2:")
print(np.round(sv.data, 3))
print("\nQubits 0 and 3 are now entangled!")
```

 **Questions:**



 1. What gates did you add for Bell measurement on qubits 1 and 2?
For the Bell measurement on qubits 1 and 2: apply CNOT(1→2) then H(1). 


 2. Qubits 0 and 3 never interacted directly. How can they become entangled?
Qubits 0 and 3 become entangled through the chain: 0 was entangled with 1, and 2 was entangled with 3. By measuring qubits 1 and 2 together in the Bell basis, the entanglement is transferred (qubit 0 becomes correlated with qubit 3 even though they never interacted.)

 3. Why is this called "entanglement swapping"?
It is called entanglement swapping because the entanglement moves: it started between (0,1) and (2,3), and after the measurement it is swapped to be between (0,3).


 4. This is the basis for quantum repeaters. Why would this be useful for long-distance quantum communication?
Over long distances, entanglement gets destroyed by noise. Quantum repeaters fix this by creating small entangled links between nearby nodes and then connecting them using entanglement swapping, allowing distant points to share entanglement without sending the qubit the whole way.
